# 모바일 장치 가격 분류(Mobile Device Price Classification)

## 데이터셋
- 파일명: pandas/data/mobile_price_classification.csv

- 컬럼 정보

|컬럼명|설명|
|---|---|
|battery_power|배터리가 한 번에 저장할 수 있는 총 에너지량, mAh 단위|
|blue|블루투스 기능 보유 여부|
|clock_speed|마이크로프로세서 실행 속도|
|dual_sim|듀얼 SIM을 지원하는지 여부|
|fc|전면 카메라 메가픽셀|
|four_g|4G 지원 여부|
|int_memory|내부 메모리(기가바이트)|
|m_dep|두께(cm)|
|mobile_wt|무게|
|n_cores|프로세서 코어 수|
|pc|메인 카메라 메가픽셀|
|px_height|픽셀 해상도 높이|
|px_width|픽셀 해상도 너비|
|ram|랜덤 액세스 메모리(RAM) 용량(메가바이트)|
|sc_h|휴대폰 화면 높이(cm)|
|sc_w|휴대폰 화면 너비(cm)|
|talk_time|배터리 한 번 충전으로 가장 오래 사용할 수 있는 시간|
|three_g|3G 지원 여부|
|touch_screen|터치스크린 지원 여부|
|wifi|wifi지원여부|
|price_range|0(저가), 1(중가), 2(고가), 3(초고가)의 값을 갖는 타겟 변수|

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

%matplotlib inline
```

```python
df = pd.read_csv('pandas/data/mobile_price_classification.csv')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='price_range', y='ram', hue='price_range')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='ram')
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(data=df, x='px_width', y='px_height')
```

```python
sns.violinplot(data=df, x='wifi', y='battery_power', hue='wifi')
```

```python
# 
sns.countplot(data=df, y='n_cores', hue='n_cores')
```
