## 1-3. 행(row) 선택하기 - 인덱스 기반

> 행을 선택하는 두 가지 핵심 접근자(accessor):
>
> | 접근자 | 기준 | 예시 |
> |--------|------|------|
> | `loc[]` | **라벨(이름)** 기반 | `df.loc[0:5]` → 0~5행 **포함** |
> | `iloc[]` | **정수 위치** 기반 | `df.iloc[0:5]` → 0~4행 (5 **미포함**) |
>
> ⚠️ **핵심 차이**: `loc`은 끝 값을 **포함**, `iloc`은 끝 값을 **미포함** (파이썬 슬라이싱과 동일)

````python
# iloc - 정수 위치 기반 인덱싱
# 첫 번째 행 (인덱스 0)
print("=== 첫 번째 행 (iloc[0]) ===")
print(df_comp.iloc[0])
print()

# 처음 3행 선택 (0, 1, 2번째 행) - 끝 값 미포함
print("=== 처음 3행 (iloc[0:3]) ===")
df_comp.iloc[0:3]
````

````python
# iloc - 행과 열을 동시에 선택 (행, 열 모두 정수 위치)
# df.iloc[행범위, 열범위]
# 처음 5행, 처음 4열 선택
print("=== iloc[0:5, 0:4] - 5행 × 4열 ===")
print(df_comp.iloc[0:5, 0:4])
print()

# 특정 행과 열을 리스트로 선택
# 0, 2, 4번째 행의 1, 3, 5번째 열
print("=== iloc[[0,2,4], [1,3,5]] - 특정 위치 선택 ===")
df_comp.iloc[[0, 2, 4], [1, 3, 5]]
````

````python
# loc - 라벨 기반 인덱싱
# loc은 인덱스 라벨과 컬럼명을 사용하여 데이터 선택
# 인덱스 0~4행, 'brand'부터 'os'까지 컬럼 선택 (끝 값 포함!)
print("=== loc[0:4, 'brand':'os'] - 라벨 기반, 끝 값 포함 ===")
df_comp.loc[0:4, 'brand':'os']
````

````python
# loc - 특정 행과 특정 컬럼을 리스트로 선택
# 인덱스 [0, 10, 100]번 행의 ['brand', 'ram_gb', 'price'] 컬럼
df_comp.loc[[0, 10, 100], ['brand', 'ram_gb', 'price']]
````

````python
# loc vs iloc 차이 비교 - 슬라이싱 끝 값 처리
print("=== loc[0:3] → 0, 1, 2, 3 (4행, 끝 값 포함) ===")
print(df_water.loc[0:3, '읍면동명':'사용량'])
print()
print("=== iloc[0:3] → 0, 1, 2 (3행, 끝 값 미포함) ===")
print(df_water.iloc[0:3, 1:4])
````

## 1-4. 조건으로 행 선택하기 (불리언 인덱싱)

> 데이터 분석에서 **가장 많이 사용**하는 데이터 선택 방법입니다.
> 
> ```python
> # 기본 문법
> df[df['컬럼'] 비교연산자 값]
> 
> # 여러 조건 결합
> df[(조건1) & (조건2)]   # AND
> df[(조건1) | (조건2)]   # OR
> df[~(조건)]             # NOT
> ```
> ⚠️ 주의: 여러 조건을 결합할 때 각 조건을 반드시 **괄호**로 감싸야 합니다.

````python
# 단일 조건 - 가격이 3000달러 이상인 컴퓨터
# 비교 연산 결과는 True/False의 불리언 Series
expensive = df_comp[df_comp['price'] >= 3000]
print(f"3000달러 이상 컴퓨터: {len(expensive):,}건 / 전체 {len(df_comp):,}건")
expensive[['brand', 'model', 'ram_gb', 'price']].head()
````

````python
# 문자열 조건 - 브랜드가 'Apple'인 컴퓨터
apple = df_comp[df_comp['brand'] == 'Apple']
print(f"Apple 컴퓨터: {len(apple):,}건")
apple[['brand', 'model', 'device_type', 'price']].head()
````

````python
# 여러 조건 AND(&) - Apple 브랜드이면서 RAM 32GB 이상
# 각 조건을 괄호로 감싸야 함
apple_high_ram = df_comp[(df_comp['brand'] == 'Apple') & (df_comp['ram_gb'] >= 32)]
print(f"Apple + RAM 32GB 이상: {len(apple_high_ram):,}건")
print(apple_high_ram[['brand', 'model', 'ram_gb', 'price']].head())
print(apple_high_ram['price'].mean().round(2))
````

````python
# 여러 조건 OR(|) - 상하수도 데이터에서 산업용 또는 영업용
industrial_or_business = df_water[(df_water['업종'] == '산업용') | (df_water['업종'] == '영업용')]
print(f"산업용 또는 영업용: {len(industrial_or_business):,}건")
industrial_or_business.head()
````

````python
# NOT(~) 조건 - 일반용이 아닌 데이터만 선택
not_general = df_water[~(df_water['업종'] == '일반용')]
print(f"일반용 제외: {len(not_general):,}건")
print(f"포함된 업종: {not_general['업종'].unique()}")
````

````python
# isin() - 여러 값 중 하나에 해당하는 행 선택 (OR 조건의 간편 버전)
# 브랜드가 Apple, Dell, HP 중 하나인 컴퓨터
target_brands = ['Apple', 'Dell', 'HP']
filtered = df_comp[df_comp['brand'].isin(target_brands)]
print(f"Apple/Dell/HP 컴퓨터: {len(filtered):,}건")
print(filtered['brand'].value_counts())
````

````python
# between() - 범위 조건 (이상 & 이하를 한 번에)
# 가격이 1000~2000 사이인 컴퓨터
mid_price = df_comp[df_comp['price'].between(1000, 2000)]
print(f"1000~2000달러 컴퓨터: {len(mid_price):,}건")
mid_price[['brand', 'model', 'price']].head()
````

## 1-5. 문자열 조건으로 선택하기 (str 접근자)

> 문자열 컬럼에 `str` 접근자를 사용하면 다양한 문자열 조건으로 필터링할 수 있습니다.
>
> | 메서드 | 설명 |
> |--------|------|
> | `str.contains('문자열')` | 특정 문자열 포함 여부 |
> | `str.startswith('문자열')` | 특정 문자열로 시작 |
> | `str.endswith('문자열')` | 특정 문자열로 끝남 |
> | `str.len()` | 문자열 길이 |

````python
# str.contains() - 모델명에 'Pro'가 포함된 컴퓨터
pro_models = df_comp[df_comp['model'].str.contains('Pro')]
print(f"'Pro' 모델: {len(pro_models):,}건")
pro_models[['brand', 'model', 'price']].head()
````

````python
# str.startswith() - 읍면동명이 '풍'으로 시작하는 지역
feng_areas = df_water[df_water['읍면동명'].str.startswith('풍')]
print(f"'풍'으로 시작하는 읍면동: {feng_areas['읍면동명'].unique()}")
print(f"건수: {len(feng_areas):,}건")
````

````python
# str.contains() + 대소문자 무시 (case=False)
# GPU 모델명에 'rtx'가 포함된 컴퓨터 (대소문자 구분 없이)
rtx_gpu = df_comp[df_comp['gpu_model'].str.contains('rtx', case=False)]
print(f"RTX GPU 컴퓨터: {len(rtx_gpu):,}건")
print(f"GPU 모델 종류: {rtx_gpu['gpu_model'].nunique()}개")
rtx_gpu['gpu_model'].value_counts().head(10)
````

## 1-6. query() 메서드로 선택하기

> `query()`는 문자열 표현식으로 조건을 작성하여, 더 읽기 쉬운 코드를 작성할 수 있습니다.
>
> ```python
> # 불리언 인덱싱 방식 (길고 복잡)
> df[(df['price'] >= 1000) & (df['brand'] == 'Apple')]
>
> # query() 방식 (짧고 간결)
> df.query("price >= 1000 and brand == 'Apple'")
> ```

````python
# query() 기본 사용 - and, or, not 키워드 사용 가능
# Dell 브랜드이면서 RAM이 32GB 이상이고 가격이 2000 이하인 컴퓨터
result = df_comp.query("brand == 'Dell' and ram_gb >= 32 and price <= 2000")
print(f"조건에 맞는 컴퓨터: {len(result):,}건")
result[['brand', 'model', 'ram_gb', 'price']].head()
````

````python
# query()에서 외부 변수 사용 - @ 기호를 붙여 참조
# 특정 가격 기준 이상인 컴퓨터 필터링
min_price = 2500
max_price = 3500
result = df_comp.query("@min_price <= price <= @max_price")
print(f"{min_price}~{max_price}달러 컴퓨터: {len(result):,}건")
result[['brand', 'model', 'price']].head()
````

````python
# query()에서 in 연산자와 한글 컬럼명(백틱) 사용
# 한글 컬럼명은 백틱(`)으로 감싸야 함
result = df_water.query("`업종` in ['산업용', '영업용'] and `사용요금` >= 100000")
print(f"산업용/영업용 중 10만원 이상: {len(result):,}건")
result.head()
````

## 1-7. 기타 유용한 선택 방법

````python
# head(), tail() - 상위/하위 N행 선택
print("=== 상위 3행 ===")
print(df_water.head(3))
print()
print("=== 하위 3행 ===")
print(df_water.tail(3))
````

````python
# nlargest(), nsmallest() - 특정 컬럼 기준 상위/하위 N행
# 가격이 가장 높은 컴퓨터 5대
print("=== 가격 TOP 5 ===")
print(df_comp.nlargest(5, 'price')[['brand', 'model', 'ram_gb', 'price']])
print()

# 사용요금이 가장 낮은 5건 (0 제외)
print("=== 사용요금 하위 5 (0 제외) ===")
df_water[df_water['사용요금'] > 0].nsmallest(5, '사용요금')
````

````python
# sample() - 랜덤 샘플링
# n: 개수 지정, frac: 비율 지정, random_state: 재현성을 위한 시드
print("=== 랜덤 5건 추출 ===")
print(df_comp.sample(n=5, random_state=42)[['brand', 'model', 'price']])
print()

# 전체의 0.01% 랜덤 추출
print("=== 전체의 0.01% 랜덤 추출 ===")
sampled = df_comp.sample(frac=0.0001, random_state=42)
print(f"추출 건수: {len(sampled)}건")
sampled[['brand', 'model', 'price']]
````

````python
# drop_duplicates() - 중복 제거 후 고유값만 선택
# 컴퓨터 데이터에서 브랜드별 고유 device_type 조합만 추출
unique_combos = df_comp[['brand', 'device_type']].drop_duplicates()
print(f"브랜드-기기유형 고유 조합: {len(unique_combos)}개")
unique_combos.sort_values(['brand', 'device_type'])
````

# 🕐 2교시: 필요 데이터 변경하기
---
## 2-1. 값 수정하기

> 데이터를 분석하기 전에 잘못된 값을 수정하거나 새로운 값을 할당해야 할 때가 많습니다.
>
> | 방법 | 문법 | 용도 |
> |------|------|------|
> | 직접 할당 | `df['컬럼'] = 값` | 전체 컬럼 수정 |
> | loc 할당 | `df.loc[조건, '컬럼'] = 값` | 조건에 맞는 행만 수정 |
> | replace() | `df['컬럼'].replace(old, new)` | 특정 값을 다른 값으로 치환 |
> | where/mask | `df['컬럼'].where(조건, 대체값)` | 조건부 값 변경 |

````python
# 실습용 데이터 복사 (원본 보존을 위해 copy() 사용)
df_c = df_comp.copy()
df_w = df_water.copy()

# loc으로 조건에 맞는 행의 특정 컬럼 값 변경
# 예: 가격이 0 이하인 행의 가격을 NaN으로 변경
print(f"변경 전 - 가격 0 이하: {(df_c['price'] <= 0).sum()}건")
df_c.loc[df_c['price'] <= 0, 'price'] = np.nan
print(f"변경 후 - 가격 NaN: {df_c['price'].isna().sum()}건")
````

````python
# replace() - 특정 값을 다른 값으로 치환
# 상하수도 데이터에서 업종명 변경
df_w['업종'] = df_w['업종'].replace({
    '일반용': '가정·일반용',     # '일반용'을 '가정·일반용'으로 변경
    '영업용': '상업·영업용'      # '영업용'을 '상업·영업용'으로 변경
})
print("변경 후 업종 목록:")
print(df_w['업종'].value_counts())
````

````python
# np.where() - 조건에 따라 다른 값 할당 (엑셀의 IF 함수와 유사)
# 가격이 2000 이상이면 '고가', 미만이면 '보통'
df_c = df_comp.copy()
df_c['가격대'] = np.where(df_c['price'] >= 2000, '고가', '보통')
print(df_c['가격대'].value_counts())
print()

# np.select() - 여러 조건에 따라 다른 값 할당
# 가격 구간별 등급 부여
conditions = [
    df_c['price'] < 1000,           # 1000 미만
    df_c['price'].between(1000, 2000),  # 1000~2000
    df_c['price'].between(2000, 3000),  # 2000~3000
    df_c['price'] >= 3000           # 3000 이상
]
choices = ['저가', '중저가', '중고가', '고가']
df_c['가격등급'] = np.select(conditions, choices, default='미분류')
print(df_c['가격등급'].value_counts())
````

## 3-1. pivot_table() - 피벗 테이블

> 피벗 테이블은 데이터를 요약하는 **교차표**를 만드는 강력한 도구입니다.
> 엑셀의 피벗 테이블과 동일한 개념입니다.
>
> ```python
> pd.pivot_table(
>     data=df,           # 데이터프레임
>     values='집계값',    # 집계할 컬럼
>     index='행',         # 행 기준
>     columns='열',       # 열 기준
>     aggfunc='mean'     # 집계 함수 (기본값: mean)
> )
> ```

````python
# 기본 피벗 테이블 - 상하수도 데이터
# 행: 읍면동명, 열: 업종, 값: 사용요금의 평균
pivot1 = pd.pivot_table(
    df_water,
    values='사용요금',       # 집계할 값
    index='읍면동명',        # 행 기준
    columns='업종',          # 열 기준
    aggfunc='mean'           # 평균 계산
).round(0)

print("=== 읍면동 × 업종별 평균 사용요금 ===")
pivot1.head(10)
````

````python
# 여러 집계 함수를 동시에 적용하는 피벗 테이블
# 컴퓨터 데이터: 브랜드 × 기기유형별 가격의 평균과 건수
pivot2 = pd.pivot_table(
    df_comp,
    values='price',
    index='brand',
    columns='device_type',
    aggfunc=['mean', 'count'],    # 여러 집계 함수
    fill_value=0                  # 결측치를 0으로 채움
).round(0)

print("=== 브랜드 × 기기유형별 가격 평균 & 건수 ===")
pivot2
````

````python
# margins=True - 소계/합계 행과 열 추가
# 사업자현황: 시도별/성별 전체 사업자 수 + 합계
pivot3 = pd.pivot_table(
    df_biz,
    values='(전체)당월',
    index='시도',
    columns='성별',
    aggfunc='sum',
    margins=True,           # 합계 행/열 추가
    margins_name='합계'      # 합계 이름 지정
)

print("=== 시도 × 성별 사업자 수 (합계 포함) ===")
pivot3
````

## 3-2. melt() - 넓은 형태를 긴 형태로 변환 (Unpivot)

> `melt()`는 **wide format → long format** 변환입니다. pivot의 반대 작업입니다.
>
> ```
> Wide format (넓은 형태):        Long format (긴 형태):
> 이름  국어  영어  수학          이름  과목  점수
> 김    90   80   70   →        김   국어   90
> 이    85   95   80             김   영어   80
>                                김   수학   70
>                                이   국어   85
>                                ...
> ```

````python
# 사업자현황 데이터가 melt의 좋은 예제
# (전체)당월, (전체)전월, (전체)전년동월 → 하나의 '기간' 컬럼으로 변환
print("=== 변환 전 (Wide 형태) ===")
print(df_biz[['업태별', '시도', '성별', '(전체)당월', '(전체)전월', '(전체)전년동월']].head(3))
print()

# melt 적용
df_biz_melted = df_biz.melt(
    id_vars=['업태별', '시도', '시군구', '성별'],   # 유지할 컬럼 (식별자)
    value_vars=['(전체)당월', '(전체)전월', '(전체)전년동월'],  # 녹일 컬럼
    var_name='기간',           # 새로 생성될 변수명 컬럼 이름
    value_name='사업자수'       # 새로 생성될 값 컬럼 이름
)

print("=== 변환 후 (Long 형태) ===")
print(f"크기: {df_biz_melted.shape}")
df_biz_melted.head(9)
````

## 3-3. stack() / unstack() - 인덱스 레벨 변환

> `stack()`과 `unstack()`은 멀티인덱스 레벨을 행↔열로 변환합니다.
>
> - **stack()**: 열 → 행 (컬럼을 인덱스 레벨로 내림)
> - **unstack()**: 행 → 열 (인덱스 레벨을 컬럼으로 올림)

````python
# 먼저 멀티인덱스 데이터를 만들기 (groupby 결과)
# 업종별 × 읍면동별 평균 요금
multi_idx = df_water.groupby(['업종', '읍면동명'])['사용요금'].mean().round(0)
print("=== 멀티인덱스 Series ===")
print(multi_idx.head(10))
print()

# unstack() - 마지막 인덱스 레벨을 컬럼으로 올림
unstacked = multi_idx.unstack(fill_value=0)
print("=== unstack() 결과 (행→열) ===")
unstacked.iloc[:, :5]  # 처음 5개 컬럼만 표시
````

````python
# stack() - 컬럼을 인덱스 레벨로 내림 (unstack의 반대)
stacked = unstacked.stack()
print("=== stack() 결과 (열→행) - 다시 원래 형태로 ===")
stacked.head(10)
````

````python
# 빈도 교차표 - 브랜드 × 기기유형 조합 건수
cross1 = pd.crosstab(df_comp['brand'], df_comp['device_type'])
print("=== 브랜드 × 기기유형 빈도 교차표 ===")
print(cross1)
print()

# normalize 옵션 - 비율로 표시
# normalize='index': 행 기준 비율, 'columns': 열 기준, 'all': 전체 기준
cross_pct = pd.crosstab(df_comp['brand'], df_comp['device_type'], normalize='index')
print("=== 브랜드별 기기유형 비율 (행 기준) ===")
(cross_pct * 100).round(1)
````

## 3-5. merge() - 데이터프레임 병합 (SQL JOIN과 동일)

> `merge()`는 **공통 키(컬럼)**를 기준으로 두 데이터프레임을 결합합니다.
> SQL의 JOIN과 동일한 개념입니다.
>
> ```python
> pd.merge(left, right, on='공통컬럼', how='inner')
> ```
>
> | how | 설명 | SQL 대응 |
> |-----|------|---------|
> | `inner` | 양쪽 모두 존재하는 키만 (기본값) | INNER JOIN |
> | `left` | 왼쪽 기준, 오른쪽에 없으면 NaN | LEFT JOIN |
> | `right` | 오른쪽 기준, 왼쪽에 없으면 NaN | RIGHT JOIN |
> | `outer` | 양쪽 모두 포함, 없으면 NaN | FULL OUTER JOIN |

````python
# merge 실습을 위한 요약 데이터프레임 생성

# 1) 읍면동별 평균 요금 테이블
df_dong_avg = df_water.groupby('읍면동명').agg(
    평균사용량=('사용량', 'mean'),
    평균요금=('사용요금', 'mean'),
    총건수=('수용가번호', 'count')
).round(1).reset_index()
print("=== 읍면동별 평균 요금 ===")
print(df_dong_avg.head())
print()

# 2) 읍면동별 업종별 건수 테이블
df_dong_type = df_water.groupby(['읍면동명', '업종']).size().reset_index(name='건수')
print("=== 읍면동별 업종 건수 ===")
print(df_dong_type.head())
````

````python
# inner merge (기본값) - 양쪽에 모두 존재하는 읍면동만 결합
merged_inner = pd.merge(df_dong_type, df_dong_avg, on='읍면동명', how='inner')
print(f"=== inner merge 결과: {len(merged_inner)}행 ===")
merged_inner.head(10)
````

````python
# left / right / outer merge 비교
# 실습을 위해 일부 데이터만 포함하는 테이블 생성

# 왼쪽: 일부 읍면동만 포함
df_left = df_dong_avg[df_dong_avg['읍면동명'].isin(['중구동', '풍산읍', '와룡면'])].copy()
# 오른쪽: 다른 일부 읍면동만 포함
df_right = df_dong_avg[df_dong_avg['읍면동명'].isin(['중구동', '와룡면', '예안면'])].copy()
df_right = df_right.rename(columns={'평균요금': '평균요금_right', '총건수': '총건수_right'})

print("왼쪽:", df_left['읍면동명'].tolist())
print("오른쪽:", df_right['읍면동명'].tolist())
print()

# 4가지 merge 비교
for how in ['inner', 'left', 'right', 'outer']:
    result = pd.merge(df_left[['읍면동명', '평균요금']], 
                       df_right[['읍면동명', '평균요금_right']], 
                       on='읍면동명', how=how)
    print(f"--- {how} merge ({len(result)}행) ---")
    print(result)
    print()
````

````python
# 키 컬럼명이 다를 때 - left_on, right_on 사용
# 컴퓨터 데이터의 브랜드별 평균 가격 테이블
df_brand_price = df_comp.groupby('brand')['price'].mean().round(0).reset_index()
df_brand_price.columns = ['제조사', '평균가격']  # 컬럼명을 한글로 변경

# '제조사'와 'brand'가 같은 의미이므로 left_on, right_on으로 매핑
merged = pd.merge(
    df_comp[['brand', 'model', 'price']].head(10),
    df_brand_price,
    left_on='brand',        # 왼쪽 DataFrame의 키 컬럼
    right_on='제조사',       # 오른쪽 DataFrame의 키 컬럼
    how='left'
)
print("=== 키 컬럼명이 다른 merge ===")
merged[['brand', 'model', 'price', '평균가격']]
````

## 3-6. concat() - 데이터프레임 연결하기

> `concat()`은 데이터프레임을 **위아래(행 방향)** 또는 **좌우(열 방향)**로 이어 붙입니다.
>
> ```python
> pd.concat([df1, df2], axis=0)  # 행 방향 (위아래)
> pd.concat([df1, df2], axis=1)  # 열 방향 (좌우)
> ```
>
> - merge는 **키(컬럼 값)**를 기준으로 결합
> - concat은 **단순히 이어 붙이기** (키 불필요)

````python
# axis=0 (기본값) - 행 방향 연결 (위아래로 이어 붙이기)
# 컴퓨터 데이터를 기기유형별로 분리 후 다시 합치기

df_desktop = df_comp[df_comp['device_type'] == 'Desktop'].head(3)
df_laptop = df_comp[df_comp['device_type'] == 'Laptop'].head(3)

print(f"Desktop: {len(df_desktop)}행, Laptop: {len(df_laptop)}행")

# 행 방향 연결
df_combined = pd.concat([df_desktop, df_laptop], axis=0, ignore_index=True)
# ignore_index=True: 인덱스를 0부터 재설정
print(f"연결 후: {len(df_combined)}행")
df_combined[['device_type', 'brand', 'model', 'price']]
````

````python
# axis=1 - 열 방향 연결 (좌우로 이어 붙이기)
# 같은 인덱스의 행끼리 옆으로 결합됨

# 브랜드별 평균 가격과 평균 RAM을 별도로 계산 후 합치기
avg_price = df_comp.groupby('brand')['price'].mean().round(0).rename('평균가격')
avg_ram = df_comp.groupby('brand')['ram_gb'].mean().round(1).rename('평균RAM')

df_brand_stats = pd.concat([avg_price, avg_ram], axis=1)
print("=== 열 방향 연결 (axis=1) ===")
df_brand_stats
````

````python
# keys 옵션 - 연결 출처를 구분할 수 있는 레벨 추가
# 여러 데이터를 합칠 때 어디서 온 데이터인지 구분 가능
df_desktop_stats = df_comp[df_comp['device_type'] == 'Desktop'].describe()[['price', 'ram_gb']]
df_laptop_stats = df_comp[df_comp['device_type'] == 'Laptop'].describe()[['price', 'ram_gb']]

# keys로 데이터 출처 표시
combined_stats = pd.concat(
    [df_desktop_stats, df_laptop_stats],
    keys=['Desktop', 'Laptop'],   # 멀티인덱스 레벨 추가
    axis=0
)
print("=== keys를 이용한 출처 구분 ===")
combined_stats
````

## 3-7. join() - 인덱스 기반 병합

> `join()`은 **인덱스**를 기준으로 병합합니다. `merge()`의 간편 버전입니다.

````python
# join() - 인덱스를 기준으로 병합
# 읍면동명을 인덱스로 설정한 두 테이블을 join

# 읍면동별 평균 사용량 (인덱스: 읍면동명)
avg_usage = df_water.groupby('읍면동명')['사용량'].mean().round(1).rename('평균사용량')

# 읍면동별 총 요금 (인덱스: 읍면동명)
total_fee = df_water.groupby('읍면동명')['사용요금'].sum().rename('총요금')

# join으로 인덱스 기반 병합
joined = avg_usage.to_frame().join(total_fee)
print("=== join() 결과 ===")
joined
````

## 3-8. 실전 종합 예제 - 3개 데이터 활용

````python
# 종합 예제 1: 컴퓨터 데이터 - 선택 + 변경 + 변형을 연속 적용
# 목표: 브랜드별 기기유형별 평균 가격 피벗 테이블 + 가격등급 추가

# 1단계: 필요한 컬럼만 선택
df_sub = df_comp[['brand', 'device_type', 'ram_gb', 'storage_gb', 'price']].copy()

# 2단계: 가격등급 컬럼 추가 (변경)
df_sub['가격등급'] = pd.cut(df_sub['price'],
                          bins=[0, 1000, 2000, 3000, float('inf')],
                          labels=['저가', '중저가', '중고가', '고가'])

# 3단계: 피벗 테이블로 변형
pivot_result = pd.pivot_table(
    df_sub,
    values='price',
    index='brand',
    columns='가격등급',
    aggfunc='count',
    fill_value=0,
    observed=True
)

print("=== 브랜드별 가격등급 분포 (건수) ===")
pivot_result
````

````python
# 종합 예제 2: 사업자현황 데이터 - melt + merge 활용
# 목표: 개인/법인을 하나의 컬럼으로 변환 후 시도별 비교

# 1단계: 개인/법인 당월 데이터만 선택
df_biz_sub = df_biz[['업태별', '시도', '성별', '(개인)당월', '(법인)당월']].copy()

# 2단계: melt로 개인/법인을 하나의 컬럼으로 변환
df_biz_long = df_biz_sub.melt(
    id_vars=['업태별', '시도', '성별'],
    value_vars=['(개인)당월', '(법인)당월'],
    var_name='사업자구분',
    value_name='사업자수'
)

# 3단계: 사업자구분 값 정리 (변경)
df_biz_long['사업자구분'] = df_biz_long['사업자구분'].replace({
    '(개인)당월': '개인',
    '(법인)당월': '법인'
})

# 4단계: 시도별 사업자구분별 합계 피벗
result = pd.pivot_table(
    df_biz_long,
    values='사업자수',
    index='시도',
    columns='사업자구분',
    aggfunc='sum'
)
result['법인비율(%)'] = (result['법인'] / (result['개인'] + result['법인']) * 100).round(1)

print("=== 시도별 개인/법인 사업자 수 ===")
result.sort_values('법인비율(%)', ascending=False)
````

| 불리언 인덱싱 | `df[df['col'] > 값]` | 조건 기반 행 선택 |
| query() | `df.query("조건식")` | 문자열 기반 조건 선택 |
| isin() | `df[df['col'].isin(리스트)]` | 여러 값 중 하나 |
| str 접근자 | `df['col'].str.contains()` | 문자열 패턴 선택 |
