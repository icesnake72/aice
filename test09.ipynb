## One-Hot Encoder

브랜드, 카테고리등 순서가 없는 데이터를 수치로 변경하면 순번이 생겨서 머신러닝 모델에 잘 못 반영될 수 있다. 순번의 의미를 없애고자 one-hot encoding을 수행한다.

```python
# pandas로 원-핫 인코딩하기
df_copy2 = df.copy()
df_copy2 = pd.get_dummies(df_copy2, columns=['brand'], dtype=int)
df_copy2.head()
```

```python
df['brand'].unique().tolist()
```

```python
# sklearn의 One-Hot Encoder 사용하기
from sklearn.preprocessing import OneHotEncoder

df_copy3 = df.copy()

oh = OneHotEncoder()
enc_brand_arr = oh.fit_transform(df_copy3['brand'].values.reshape(-1,1)).toarray()
print(enc_brand_arr)
enc_brand_arr = enc_brand_arr.astype('i1')

# oh.categories_는 OneHotEncoder가 학습(fit) 후 저장하는 속성으로, 
# 각 컬럼별로 발견한 카테고리 목록을 담고 있습니다.
cols = [ 'brand_'+v for v in oh.categories_[0]]
df_oh = pd.DataFrame(enc_brand_arr, columns=cols)
df_ohenc_lb = pd.concat([df_copy3, df_oh], axis=1)
df_ohenc_lb.head()
```
